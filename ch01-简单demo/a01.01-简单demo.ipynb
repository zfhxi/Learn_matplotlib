## 1.简单的y=x

```python
import matplotlib.pyplot as plt

x = [1, 0]
y = [1, 0]
plt.figure()
plt.plot(x, y)
plt.show()
```

## 2.简单的折线图

```python
x = [1, 2, 3, 4]
y = [1.2, 2.5, 4.5, 7.3]

# 设定线条样式，颜色洋红，标记点为x号，线条风格虚线，宽度为4
plt.plot(x, y, color='m', marker='x', linestyle=':', linewidth=4)
# 保存图片，设定分辨率8x120=960
# plt.savefig("img/折线图demo.png", dpi=120)
# 显示图片
plt.show()
```

## 3.画y=sin(x)

生成x数据时有时也用`np.linspace`，`np.logspace`，区别见<https://blog.csdn.net/duanlianvip/article/details/100102797>

```python
import numpy as np

x = np.arange(-5, 5, 0.02)
y = np.sin(x)

# 坐标控制，x为[-pi,pi]，y为[-2,2]，即plt.axis([x_min, x_max, y_min, y_max])
plt.axis([-np.pi, np.pi, -2, 2])
plt.plot(x, y, color='b', linestyle='-.', linewidth=1)
plt.show()
```
