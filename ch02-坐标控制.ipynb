```python
import matplotlib.pyplot as plt
import numpy as np
```

以y=sin(x)为例

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.02)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## 1.坐标范围设置

```python
plt.plot(x,y)
plt.axis([-10, 10, -2, 2])
plt.show()
```

## 2.坐标的间隔

```python
# x坐标的间隔
plt.plot(x,y)
plt.xticks([i*np.pi/2 for i in range(-4, 5)],
           [str(i*0.5)+"$\pi$" for i in range(-4, 5)])
# y坐标的间隔
plt.yticks([i*0.5 for i in range(-4, 5)])
plt.show()
```

## 3.坐标的标记

```python
plt.plot(x,y)
# x坐标的标记
plt.xlabel("x")
# y坐标的标记
plt.ylabel("y")
plt.show()
```
